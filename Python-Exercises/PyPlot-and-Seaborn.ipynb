```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

df = pd.read_csv(r"https://raw.githubusercontent.com/babdelfa/class/main/EmployeeSalaries.csv")  
print(df.head())
print(df.columns)
```

```python
sns.catplot(x = 'MBA', y = 'Salary', data=df,kind="box")
plt.title(label="Salary Comparison (MBA vs Non-MBA)")
```

```python
sns.jointplot( x='Age' , y='Salary', data=df)
```

```python
sns.stripplot(x='MBA',y='Salary', data=df, hue='MBA' )
plt.title(label = "Salary Comparison by MBA Status")
```

```python
fg, ax = plt.subplots(figsize = (5,3))
plt.bar(df["MBA"].unique(),df["MBA"].value_counts())
plt.xlabel(xlabel= "MBA Status")
plt.ylabel(ylabel= "Frequency")
plt.title (label = "MBA vs Non-MBA Employees")
```

```python
fg, ax = plt.subplots (2,2, figsize=(8,6))
ax[0,0].bar(df["MBA"].unique(),df["MBA"].value_counts())
ax[0,0].set_ylabel('Frequency')
ax[0,0].set_title('MBA vs Non-MBA Employees')

ax[0,1].plot(df['Age'],df['Salary'], 'o')
ax[0,1].set_xlabel('Age')
ax[0,1].set_ylabel('Salary')
ax[0,1].set_title('Age vs Salary')

Non_MBA = df.loc[df['MBA']=='No']
ax[1,0].hist(Non_MBA['Salary'])
ax[1,0].set_xlabel('Salary')
ax[1,0].set_ylabel('Frequency')
ax[1,0].set_title('Non-MBA Salary Distribution')

MBA = df.loc[df['MBA']=='Yes']
ax[1,1].hist(MBA['Salary'])
ax[1,1].set_xlabel('Salary')
ax[1,1].set_ylabel('Frequency')
ax[1,1].set_title('MBA Salary Distribution')

plt.tight_layout()
plt.show()
```

```python
fg, ax = plt.subplots (2,2, figsize=(8,6))
ax[0,0].bar(df["MBA"].unique(),df["MBA"].value_counts())
ax[0,0].set_ylabel('Frequency')
ax[0,0].set_title('MBA vs Non-MBA Employees')

sns.scatterplot(data = df , x = 'Age', y = 'Salary', hue = 'MBA', ax = ax[0,1])

Non_MBA = df.loc[df['MBA']=='No']
ax[1,0].hist(Non_MBA['Salary'])
ax[1,0].set_xlabel('Salry')
ax[1,0].set_ylabel('Frequency')
ax[1,0].set_title('Non-MBA Salary Distribution')

MBA = df.loc[df['MBA']=='Yes']
ax[1,1].hist(MBA['Salary'])
ax[1,1].set_xlabel('Salary')
ax[1,1].set_ylabel('Frequency')
ax[1,1].set_title('MBA Salary Distribution')

plt.tight_layout()
plt.show()
```
